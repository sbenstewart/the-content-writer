```python
def f(x, y):
    return np.power(y, x) - 1

x = np.linspace(0, 10, 100)
y = np.linspace(0, 2, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
ax = plt.axes(projection='3d')
plt.tight_layout()
#ax.view_init(15, 195)
ax.ticklabel_format(useOffset=False, style='plain')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')

ax.set_xlabel('Number of Fingers', labelpad=10)
ax.set_ylabel('Number of Positions for each Finger', labelpad=10)

#ax.set_zscale('log')
ax.set_zlabel('Max number that can be counted', labelpad=10);

plt.xticks(np.arange(0, 11, step=2))
plt.yticks(np.arange(0, 3, step=1))

#xticks([0, 1, 2], ['January', 'February', 'March'], rotation=20)
```
