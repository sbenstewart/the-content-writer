```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats 
import scipy.stats as ss
%matplotlib inline
```

```python
img = cv2.imread('flatten-the-curve-#5.png')
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img)
```

```python
blur = img
```

```python
for i in range(0,10):
    blur = cv2.GaussianBlur(blur,(45,45),0)
```

```python
plt.figure(figsize=(10,10))
plt.hist(blur.ravel(),  histtype="stepfilled")
plt.hist(img.ravel(),25,[0,256], histtype="stepfilled")
#'bar', 'barstacked', 'step', 'stepfilled'
#256,[0,256],
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(blur)
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(10,8)})
sns.set_style("white")
sns.set_context("talk")
sns.despine()
x = img.ravel()
ax = sns.distplot(x, label = 'histogram for the unblurred content', color='red')
x = blur.ravel()
ax = sns.distplot(x, label = 'histogram for the blurred content', color='blue')
plt.legend(bbox_to_anchor=(0.6, 1), loc=1)
```
